> Add # sign in front of phrase to create heading. There are six heading levels from H1 to H6

```
# Heading level 1 
## Heading level 2 
### Heading level 3
#### Heading level 4 
##### Heading level 5
###### Heading level 6
```




# Blank line

> Use a blank line to separate one or more lines of text. (like TeX)

I really like using Markdown.

I think I'll use it from now on.

> End a line with two or more spaces, and then type Return/Enter

I really like using Markdown.  
I think I'll use it from now on.

# Emphasis

## Bold
> Add 2 `*` or 2 `_` before **AND** after a word or phrase.

I love **bold text**.

I love __bold text__.

Love **is** bold

## Italic
> Add 1 `*` or 1 `_` before **AND** after a word or phrase.

I love *bold text*.

I love _bold text_.

Love *is* bold

## Bold & Italic
> Add 3 `*` or 3 `_` before **AND** after a word or phrase.

I love ***bold text***.

I love ___bold text___.

Love ***is*** bold

## Strikethrough
> Add `~~` before **AND** after a word

The world is ~~flat~~ round.

# Blockquotes

## Single block-quote
> Add a `>` in front of a paragraph.

> Dorothy followed her through many rooms


## Blockquotes with Multiple Paragraphs
> Add a `>` on the blank lines between the paragraphs.

> This the first paragraph.
>
> And this is the second paragraph.

# Links/URLs

> Full-version: Write as `[Link text](Link URL "Title")`. Title is hover text of link

Use [Duck Duck Go](https://duckduckgo.com "My search engine!").



> Short-verson: Write as `[Link text]{Link URL}`

Use [Duck Duck Go](https://duckduckgo.com).



## URLs and Email
To quickly turn a URL or email address into a link, enclose it in angle brackets

<https://eff.org>

<fake@example.com>
